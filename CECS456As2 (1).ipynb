```python
import numpy as np
import matplotlib.pyplot as plt

#Question 1a
def graph1a():
    #set y value to log2 with a base of 10
    yValue = -np.log10(2)
    #set values for stationary point's X and Y
    statX = [0.5]
    statY = [yValue]
    plt.xlim(0, 1)
    plt.ylim(-0.4, 0.1)
    plt.grid()
    #plot the stationary point on the graph
    plt.plot(statX, statY, marker="o", markersize=5, markeredgecolor="green", markerfacecolor="green")

    #set values for w to be graphed in gw
    w = np.linspace(0.01, 0.99, 100)

    #gw equation
    gw = w * np.log10(w) + (1-w) * np.log10(1-w)
    #graph the equation onto the plot
    plt.plot(w, gw)
    plt.xlabel("w")
    plt.ylabel("g(w)")
    plt.show()

graph1a()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Question 1b
def graph1b():
    #changes the plot to a 3d graph
    ax = plt.axes(projection = "3d")
    ##set values for statpoint mentioning w1, w2, and g(w)
    statW1 = [(-2/5)]
    statW2 = [(-1/5)]
    statY = [(-3/10)]
    #plot a point
    ax.scatter(statW1, statW2, statY, s = 100, color="blue", marker="o")

    #set values for w1 and w2
    #set w1 and w2 as meshgrids?
    w1 = np.linspace(-1, 1, 100)
    w2 = np.linspace(-1, 1, 100)
    w1, w2 = np.meshgrid(w1, w2)

    # graphing equation 
    gw = (1/2) * ((2 * w1 ** 2) + (2 * w1 * w2) + (3 * w2 ** 2)) + w1 + w2
    # set labels for x, y, and z
    ax.set_xlabel("w1")
    ax.set_ylabel("w2")
    ax.set_zlabel("g(w)")
    # set limits for the graph dimensions
    ax.set_xlim(-1, 1)
    ax.set_ylim(-1, 1)
    ax.set_zlim(-2, 6)
    plt.title("3D graph of g(w)")
    #plot the equation onto the 3d map
    ax.plot_surface(w1, w2, gw, color ="green", alpha= 0.3)
    plt.show()
    
graph1b()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Question 2
def gd():
    # def g(w):
    #     return (1 / 50) * (w ** 4 + w ** 2 + 10 * w)

    def derivative(w):
        return (1/50) * (4 * w ** 3 + 2*w +10)
    # Set the initial value for w0
    w0 = 2
    # Set iteration counter
    iterations = 1000

    # Set an array for step lengths (alphas)
    alphas = [1, 1e-1, 1e-2]

    # Loop over different step sizes (alphas)
    for i in range(len(alphas)):
        # Initialize w to the initial value
        w = w0
        iterationArr = []
        weightArr = []
        # Loop for a fixed number of iterations
        for j in range(iterations):
            iterationArr.append(j)
            weightArr.append(w)
            # Calculate the gradient using np.gradient
            gradient = derivative(w)
            # Update w using the gradient and alpha
            w = w - alphas[i] * gradient
        #plot the equation 
        plt.plot(iterationArr, weightArr, label=f'alpha = {alphas[i]}')
    plt.title("Cost Function History")
    plt.xlabel("Iterations")
    plt.ylabel("Cost Evaluation (g(w))")
    #show all the alpha values being plotted
    plt.legend()
    plt.show()

gd()
```

I believe that the step length value works best when alpha = 0.1 because this line goes to the lowest point and gradually falls toward the minimum cost. While alpha = 1 is too steep and alpha = 0.01 never reaches the minimum.
